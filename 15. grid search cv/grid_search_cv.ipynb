```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
import pandas as pd
import numpy as np
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
df = pd.DataFrame(digits.data,columns=digits.feature_names)
df['target'] = digits.target
df
```

```python
x_axis = df.drop(['target'],axis='columns')
x_axis
```

```python
y_axis = digits.target
y_axis
```

## Approach 1: Use train test split and manually tune parameters by trial and error

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x_axis,y_axis,test_size=0.2,random_state=42)
```

```python
svm = SVC(kernel='rbf',C=10,gamma='auto')
svm.fit(X_train,y_train)
svm.score(X_test,y_test)
```

```python
svm1 = SVC(kernel='linear',C=10,gamma='auto')
svm1.fit(X_train,y_train)
svm1.score(X_test,y_test)
```

## Approach 2 : use k fold validation

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(LogisticRegression(),x_axis,y_axis,cv =5)
```

```python
cross_val_score(DecisionTreeClassifier(max_depth=5,random_state=42),x_axis,y_axis,cv=5)
```

```python
cross_val_score(RandomForestClassifier(n_estimators=10),x_axis,y_axis,cv=5)
```

```python
cross_val_score(SVC(gamma='auto',kernel='rbf',C=50),x_axis,y_axis,cv=5)
```

```python
cross_val_score(GaussianNB(),x_axis,y_axis,cv=5)
```

```python
cross_val_score(MultinomialNB(),x_axis,y_axis,cv=5)
```

## Above approach is very manual. We can use for loop as an alternative

```python
kernel = ['rbf' , 'linear']
c = [1,10,20,50]
avg_score = {}
for kval in kernel:
    for cval in c:
        cv_score = cross_val_score(SVC(kernel=kval,C=cval,gamma='auto'),digits.data,digits.target,cv=5)
        avg_score[kval + '_' + str(cval)] = np.average(cv_score)

avg_score
```

## Approach 3 : Use grid search cv

```python
from sklearn.model_selection import GridSearchCV
clf = GridSearchCV(SVC(gamma='auto'),{
    'C' : [1,10,20,50],
    'kernel' : ['linear','rbf']
},cv=5,return_train_score=False)
clf.fit(digits.data,digits.target)
clf.cv_results_
```

```python
df2 = pd.DataFrame(clf.cv_results_)
df2
```

```python
clf.best_params_
```

```python
df2[['param_C', 'param_kernel', 'mean_test_score']]
```

```python
clf.best_score_
```

```python
from sklearn.model_selection import RandomizedSearchCV
rs = RandomizedSearchCV(SVC(gamma='auto'),{
    'C' : [1,10,20,50],
    'kernel' : ['linear','rbf']
},cv=5,return_train_score=False,n_iter=3)
rs.fit(digits.data,digits.target)
rs.cv_results_
df3 = pd.DataFrame(rs.cv_results_)
df3[['param_C', 'param_kernel', 'mean_test_score']]
```

```python
rs.best_params_
```

```python
rs.best_score_
```

## Different Models with different parameters

```python
model_params = {
    'svm3' : {
        'model' : SVC(gamma='auto'),
        'params' : {
            'C' : [1,10,20,30],
            'kernel':['linear','rbf']
    }},
    'logistic_reg' : {
        'model' : LogisticRegression(),
        'params' : {
            'C' : [10,1,20,'auto']
        }
    },
    'Decision_tree' : {
        'model': DecisionTreeClassifier(),
        'params' : {
            'max_depth' : [5,10,20,30],
            'criterion' : ['gini','entropy']
        }
    },
    'Random_forest':{
        'model' : RandomForestClassifier(),
        'params' : {
            'max_depth' : [5,10,20,30]
        }
    },
    'Multinomial': {
        'model': MultinomialNB(),
        'params' : {}
    },
    'Gaussian': {
        'model': GaussianNB(),
        'params' : {}
    }
}
```

```python
scores3 = []

for model_name,mp in model_params.items():
    clf = GridSearchCV(mp['model'],mp['params'],cv=5,return_train_score=False)
    clf.fit(digits.data,digits.target)
    scores3.append({
        'model' : model_name,
        'best_score': clf.best_score_,
        'best_params': clf.best_params_
    })

df4 = pd.DataFrame(scores3 , columns=['model' , 'best_score', 'best_params'])
df4
```
